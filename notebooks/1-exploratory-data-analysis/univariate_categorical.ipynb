# Univariate EDA

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Preparation

```python
data_df = pd.read_csv("../data/sample.csv")
feature_analysis_df = pd.read_csv("../data/column_nature.sample.csv")
```

## Helper Functions

```python
def separator(text: str = ""):
    if text != "":
        text = " " + text + " "

    text_to_print = (
        "=============="
        + text
        + "============================================================================================"
    )

    if text != "":
        text_to_print = text_to_print[: -len(text)]
    print(text_to_print)
```

```python
def non_graphical_categorical_data_eda(df: pd.DataFrame, column_name: str):
    # Non-graphical EDA
    separator("Non-graphical EDA for column: " + column_name)
    value_count = df[column_name].value_counts().sort_index()

    separator("Count for each catogory")
    display(pd.DataFrame(value_count))
    
def graphical_categorical_data_eda(df: pd.DataFrame, column_name: str):
    # Non-graphical EDA
    separator("Non-graphical EDA for column: " + column_name)
    value_count = df[column_name].value_counts().sort_index()
    value_count_proportion = df[column_name].value_counts(normalize=True).sort_index()
    category_list = list(value_count.index)

    separator("Count for each catogory")
    display(pd.DataFrame(value_count))
    separator("Proportion for each catogory")
    display(pd.DataFrame(value_count_proportion))

    sum_of_count = 0
    for cat in category_list:
        sum_of_count += value_count[cat]
    sum_of_count_is_equal = df.shape[0] == sum_of_count
    print("No missing values\t: ", sum_of_count_is_equal)

    sum_of_proportions = 0
    for cat in category_list:
        sum_of_proportions += value_count_proportion[cat]
    sum_of_proportions = round(sum_of_proportions, 5)
    proportions_is_1 = 1 == sum_of_proportions
    if proportions_is_1:
        print("Total proportion is 1\t: ", proportions_is_1)
    else:
        print(
            "Total proportion is 1\t: ",
            proportions_is_1,
            ", Actual: ",
            sum_of_proportions,
        )

    # Visualization
    value_count.plot.bar()
    value_count_proportion.plot.bar()
    plt.show()
```

## Iterate Categorical Features and Produce Analysis

```python
categorical_features_df = feature_analysis_df.loc[
    feature_analysis_df["columnNature"].isin(["categorical", "nominal", "ordinal", "binary"])
]

for ix, row in categorical_features_df.iterrows():
    
    column_name = row["columnName"]
    
    if column_name in data_df.columns.to_list():
        graphical_categorical_data_eda(data_df, column_name)
        print()
```

## Alive only EDA

```python
alive_data = data_df[data_df['Outcome'] != 'Dead'].copy()
alive_data.shape
```

```python

for ix, row in categorical_features_df.iterrows():
    
    column_name = row["columnName"]
    
    if column_name in alive_data.columns.to_list():
        non_graphical_categorical_data_eda(alive_data, column_name)
        print()
```

## Dead only EDA

```python
alive_data = data_df[data_df['Outcome'] == 'Dead'].copy()
alive_data.shape
```

```python

for ix, row in categorical_features_df.iterrows():
    
    column_name = row["columnName"]
    
    if column_name in alive_data.columns.to_list():
        non_graphical_categorical_data_eda(alive_data, column_name)
        print()
```
